```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
nltk.download('stopwords')
```

```python
df  = pd.read_csv(r"C:\Users\DELL\Downloads\archive (5)\customer_support_tickets.csv")
df
```

```python
stop_words = set(stopwords.words('english'))
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)  
    return text
df['text'] = (
    df['Ticket Subject'] + " " +
    df['Ticket Description'] + " " +
    df['Product Purchased'].astype(str)
)
df['clean_text'] = df['text'].apply(clean_text)
```

TEXT CLEANING

```python
df.head()
```

```python
vectorizer = TfidfVectorizer(
    max_features=10000,
    ngram_range=(1,2),   
    min_df=2,
    max_df=0.9
)

X = vectorizer.fit_transform(df['clean_text'])
```

FEATURE EXTRACTION

```python
y_category = df['Ticket Type']
```

TICKET CATEGORY CLASSIFICATION

```python
y_priority = df['Ticket Priority']
```

PRIORITY PREDICTION

```python
X_train, X_test, y_cat_train, y_cat_test = train_test_split(
    X, y_category, test_size=0.2, random_state=42)

_, _, y_pri_train, y_pri_test = train_test_split(
    X, y_priority, test_size=0.2, random_state=42)

model_cat = LogisticRegression()
model_cat.fit(X_train, y_cat_train)

model_pri = LogisticRegression()
model_pri.fit(X_train, y_pri_train)

y_cat_pred = model_cat.predict(X_test)
y_pri_pred = model_pri.predict(X_test)

print("Category Accuracy:", accuracy_score(y_cat_test, y_cat_pred))
print("\nCategory Classification Report:\n")
print(classification_report(y_cat_test, y_cat_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_pri_test, y_pri_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Priority Prediction")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
